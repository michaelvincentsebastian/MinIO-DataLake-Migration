```python
import duckdb
import pandas as pd
```

```python
con = duckdb.connect()
con.sql("INSTALL ducklake; LOAD ducklake;")
```

```python
targetPathCSV = "./demoData/dataJakarta.csv"
targetPathParquet = "./demoData/dataJakarta.parquet"
catalogMetadata = './catalog/catalog.ducklake'
managedStoredData = './managedStoreData/'
```

```python
con.sql(f"ATTACH 'ducklake:{catalogMetadata}' AS LocalLake (DATA_PATH '{managedStoredData}');")
con.sql("USE LocalLake;")
```

CSV Version

```python
try:
    con.sql(f"""
        COPY staging.dataJakarta 
        TO '{targetPathCSV}' 
        (FORMAT CSV, HEADER);
    """)
    print("✅ Done Export to CSV")
except Exception as e:
    print(f"❌ Terjadi Error: {e}")
```

Versi Parquet

```python
try:
    con.sql(f"""
        COPY staging.dataJakarta 
        TO '{targetPathParquet}' 
        (FORMAT PARQUET);
    """)
    print("✅ Done Export to Parquet")
except Exception as e:
    print(f"❌ Terjadi Error: {e}")
```
