```python
import pandas as pd
import duckdb
import boto3
```

```python
con = duckdb.connect()
```

```python
parquet = 'staging/parquet/duckdb/sumateraSelatan/palembang.parquet'
csv = 'staging/csv/duckdb/sumateraSelatan/palembang.csv'
```

```python
minioAuth = {
    'accessKey': 'minioadmin',
    'secretKey': 'miniopassword',
    'bucket': 'data-dapodik-kemendikdasmen',
    'endpoint': 'http://localhost:9000'
}

minioEndPoint = minioAuth['endpoint'].replace('http://', '')

# MinIO Setup for DuckDB
con.sql(f"SET s3_endpoint = '{minioEndPoint}';") 
con.sql(f"SET s3_access_key_id = '{minioAuth['accessKey']}';")
con.sql(f"SET s3_secret_access_key = '{minioAuth['secretKey']}';")
con.sql("SET s3_use_ssl = false;")
con.sql("SET s3_url_style = 'path';")

# MinIO Setup for boto3
s3_client = boto3.client(
    's3',
    endpoint_url = minioAuth['endpoint'],
    aws_access_key_id = minioAuth['accessKey'],
    aws_secret_access_key= minioAuth['secretKey'],
    verify = False
)
```
