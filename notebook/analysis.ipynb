```python
import duckdb
```

```python
minioAuth = {
    'accessKey': 'minioadmin',
    'secretKey': 'miniopassword',
    'bucket': 'data-dapodik-kemendikdasmen',
    'endpoint': 'http://localhost:9000'
}
```

```python
currentMiniofilePath = f"s3://{minioAuth['bucket']}/JawaBarat/dataDKIJakarta.parquet"
```

```python
con = duckdb.connect()
con.sql("INSTALL httpfs; LOAD httpfs;")
```

```python
con.sql(f"SET s3_endpoint = '{minioAuth['endpoint'].replace('http://', '')}';")
con.sql("SET s3_use_ssl = false;") 
con.sql(f"SET s3_access_key_id = '{minioAuth['accessKey']}';")
con.sql(f"SET s3_secret_access_key = '{minioAuth['secretKey']}';")
con.sql("SET s3_url_style = 'path';")
```

```python
try:
    print("\n--- TEST: Mengkueri File Parquet Langsung dari MinIO ---")
    # Coba SELECT sederhana pada file LIVE Anda
    test_query = con.sql(f"SELECT COUNT(*) FROM read_parquet('{currentMiniofilePath}') LIMIT 1;").fetchone()
    
    if test_query:
        print(f"‚úÖ TEST SUKSES! File Parquet dapat dibaca. COUNT: {test_query[0]}")
        print("Lanjutkan ke eksekusi kode DuckLake...")
    
except Exception as e:
    print(f"\n‚ùå TEST GAGAL! File TIDAK BISA dibaca. Error: {e}")
    print("\nüí° Aksi yang harus dilakukan:")
    print("1. Cek kembali di GUI MinIO (Port 9000) apakah file 'dataDKIJakarta.parquet' benar-benar ada di folder 'JawaBarat'.")
    print("2. Jika ada, coba UNDUH file tersebut dan pastikan bisa dibuka di Pandas lokal. Jika tidak bisa dibuka, file tersebut rusak.")
    print("3. Coba upload ulang file Parquet ke MinIO menggunakan boto3 atau DuckDB COPY TO.")
    
    # Hentikan eksekusi script utama jika test gagal
    raise
```
